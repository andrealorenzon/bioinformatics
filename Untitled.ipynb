```r
library(Biostrings)
library(stringi)

word.size = 6
interval = word.size - 1
output = 'word_list_real.txt'
rand.output = 'word_list_random.txt'

# Function to count words in Biostring object, it returns a data.frame
count.words = function(x) {
  string = toString(x)
  seqid = names(x)
  word = substring(string, 1:(width(string)-interval),
                              word.size:width(string))
  data = data.frame(word = word)
  data$position = 1:nrow(data)
  data$seqid = rep(seqid,nrow(data))
  data
}

# Function to count words after randomizing the sequence maintaining the same composition, shuffling the nucleotides
rand.count.words = function(x) {
  string = toString(x)
  string = stri_rand_shuffle(string)
  seqid = names(x)
  word = substring(string, 1:(width(string)-interval),
                   word.size:width(string))
  data = data.frame(word = word)
  data$position = 1:nrow(data)
  data$seqid = rep(seqid,nrow(data))
  data
}

# Read the fasta with the promoters -100 to 0 extracted from Biomart
seq = readDNAStringSet('dmelanogaster_promoter_100_to_0.fa')

# Count the words for every promoter, writes the output to an external txt file
for(i in 1:length(seq)) {
  data = count.words(seq[i])
  write.table(data, 
              output, append=T,
              col.names = F, 
              row.names = F, 
              quote = F)
}

# Load the written txt file
data = read.table(output, header = F)
head(data)

# Count the occurrences and select the top 15 words
counts = table(data$V1)
top = head(sort(counts, decreasing=T), 15)
par(mfrow=c(3,5))
top.words = names(top)

# Open the graphilcal window to plot 15 histograms
par(mfrow=c(3,5))

# Plot the histograms for the top 15
for(i in top.words) {
  hist(subset(data, V1 == i)$V2,
       breaks=100, main=i, ylim=c(0,150))
}

# Generate the random sequences and count the occurrences of words writing to a file
for(i in 1:length(seq)) {
  data = rand.count.words(seq[i])
  write.table(data, 
              rand.output, append=T,
              col.names = F, 
              row.names = F, 
              quote = F)
}

# Read the random counts and plot the top 15
rand = read.table(rand.output, header = F)
rand.counts = table(rand$V1)
rand.top = head(sort(rand.counts, decreasing=T), 15)
par(mfrow=c(3,5))
for(i in top.words) {
  hist(subset(rand, V1 == i)$V2,
       breaks=100, main=i, ylim=c(0,150))
}

```
